```python
import garpar as gp
import numpy as np
import attr
```

```python
df = gp.make_market(stock_number=3)
```

```python
@attr.s(repr=False)
class Market:
    _df = attr.ib()
    
    def __repr__(self):
        return repr(self._df)
    
    def _repr_html_(self):
        return self._df._repr_html_()
    
    def returns(self):
        df = self._df.copy()
        
        price_columns = df.columns[2:]
        
        prices = df[price_columns]
        returns = prices.pct_change()
        
        df[price_columns] = returns
        return df
    
    def markowikz(self):
        returns = self.returns()
        


market  = Market(df=df)

returns = market.returns()
returns.describe()
```
